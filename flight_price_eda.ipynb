#### EDA and Feature Engineering Flight Price Prediction


#### Features
##### The various features of the cleaned dataset are explained below:
1. Airline : The name of the airline company is stored in the airline column. it is a categorical feature having 6 different airlines
2. Flight: Flight stores information regarding the plane's flight code. it is a categorical feature
3. Source city: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4. Departure time: This is a derived categorical feature obtained created by grouping time period in bins. it stores information about the departure time and have 6 unique time labels
5. Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6. Arrival time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels with information about the arrival time.
7. Desitnation city: City where the flight will land. it is a categorical feature having 6 unique cities.
8. Class: A categorical feature that contains information on seats class. it has two distinct values, Business and economy.
9. Duration: A continious feature that displays the overall time it takes to travel between cities in hours
10. Days left: It is a derived characteristic that is calculated by substrcting the trip date by the booking date.
11. Price: Target variable stores the information of the ticket price.

```python
## Feature engineering
### Hamdling date of journey (object) separate it to date month and year

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
### Even though we have splitted the date of journey to date, month and year
## The data type of date month and year are still objects 
## Let's make them numeric values
## Typecasting
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
## Dropping Date of journey  column
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
## Removing extra strings from  the  arrival time
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
## Getting arrival hour and time (numerical) from Arrival Time (object)
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour']  =  df['Arrival_hour'].astype(int)
df['Arrival_min']  =  df['Arrival_min'].astype(int)
```

```python
## Dropping Arival_Time column since it is no longer required
df.drop('Arrival_Time',axis=1,inplace=True)
```
